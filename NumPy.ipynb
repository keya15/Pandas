### Boolean Indexing

```python
names = np.array(['bob', 'joe', 'will', 'bob', 'will', 'joe', 'joe'])
```

```python
data = np.random.randn(7,4)
```

```python
names
```

```python
data
```

```python
names == 'bob'
```

```python
data[names == 'bob']
```

```python
data[names == 'bob', 2:]
```

```python
data[names == 'bob', 3]
```

```python
names != 'bob'
```

```python
data[~(names == 'bob')]
```

```python
data[data < 0] = 0
```

```python
data
```

```python
data[names != 'joe'] = 7
```

```python
data
```

### Fancy Indexing

```python
arr = np.empty((8,4))
```

```python
arr
```

```python
for i in range(8):
    arr[i] = i
```

```python
arr
```

```python
arr[[-3,-5,-7]]
```

```python
arr = np.arange(32).reshape((8,4))
```

```python
arr
```

```python
arr[[1, 5, 7, 2],[0, 3, 1, 2]]
```

```python
arr[[1,5,7,2]][:, [0,3,1,2]]
```

### Transposing arrays and swapping axes

```python
arr = np.arange(15).reshape((3,5))
```

```python
arr
```

```python
arr.T
```

```python
arr = np.random.randn(6,3)
```

```python
np.dot(arr.T, arr)
```

```python
arr = np.arange(16).reshape((2,2,4))
```

```python
arr
```

```python
arr.transpose((1,0,2))
```

```python
arr
```

```python
arr.swapaxes(1,2)
```
